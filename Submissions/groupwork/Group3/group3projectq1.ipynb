```python

#Set a number, default parameters
secret_number = 50
guess = 0
attempt = 0
#Ask the user for a number in a continious loop till the condions are met.
while guess != secret_number:
      try:
       attempt = attempt + 1
       guess = int(input("Guess the secret number"))
       if guess > secret_number:
        #Displays too low when the number is smaller
          print("Too high")
        #Displays too high when the number is larger
       elif guess < secret_number:
          print("Too low")
      #displays the correct when the required number is guessed
       else :
              print("Correct")

       print(f"You were successful in {attempt} Attempts"])

      except ValueError:
         print("Please type in a number")
```
