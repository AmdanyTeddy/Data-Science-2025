```python
#a program that asks for the name age and a number from the user and uses conditional
#statements to output whether a number is positive negative or zero

#handle a value error
try:
# Asks the user's name and age
  name = input("Enter your name : ")
  age = int(input("Enter your age : "))

# prints formatted greeting using an f-string
  print(f"Hello {name} welcome to this session. You are {age} years of age")

# Asks the user to input a number
  number = int(input("Enter a number : "))

# uses conditional statements to output whether the number is positive, negative or zero
  if number > 0:
    print("The number is positive")

  elif number < 0:
    print("The number is negative")

  elif number == 0:
    print("The number is zero")
except ValueError:
  print("please enter the correct data type")
```
