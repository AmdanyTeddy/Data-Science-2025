```python
#initialize 3 variables
secret_number=7
guess=0
attempt=0

#handle value errors using try except
try:

#define a while loop that Repeats until the user guesses correctly.
  while guess!=secret_number:

    #handle keyboardinterupt error using try except
    try:

       #increment attempt so as to record the number of attempts made by the user
      attempt=attempt+1

      #ask the user to guess a number
      guess=int(input("Guess a number:"))

      #define an if elif else statement to Compares the guess to the secret and prints:
      # “Too low” if the guess is less than the secret.
      # “Too high” if the guess is greater.
      # “Correct!” if the guess matches.
      if guess<secret_number:
        print("Too low")
      elif guess>secret_number:
        print("Too high")
      else:
        print("Correct!")

    except ValueError:
        print("Invalid input. Please enter a valid integer.")

  #print the number of attempts the user made
  print(f"Attempts= {attempt}")
except KeyboardInterrupt:
        print("\nProgram terminated by the user.")
```

```python

#Prompts the user to enter a full sentence.
sentence=input(f"enter a sentence:")

#Number of characters (excluding spaces)
characters=len(sentence.replace(" ",""))

#Number of words.
#split- divide a string into a list of substrings.
#len-number of words
words=len(sentence.split())

#The sentence in uppercase and lowercase.
uppercase=sentence.upper()
lowercase=sentence.lower()


#The sentence reversed.
reversed=sentence[::-1]

#Outputs a formatted report using f-strings and aligned indentation.
print(f"sentence: {sentence}")
print(f"Number of characters: {characters}")
print(f"Number of words: {words}")
print(f"Uppercase: {uppercase}")
print(f"Lowercase: {lowercase}")
print(f"Reversed: {reversed}")






```
