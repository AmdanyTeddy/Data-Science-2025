```python
# Challenge 1: Counting Vowels in a Sentence
# Description
# Write a function that counts how many vowels (a, e, i, o, u, case insensitive) appear in a given sentence.
# Roles
# •	Coder: Implement count_vowels(sentence) using a for loop.
# •	Tester: Prepare ≥5 test cases (empty string, no vowels, all vowels, mixed case).
# •	Presenter: Demonstrate the function on sample inputs and explain the loop logic.

#create a function that accepts user input as its parameter (sentence)
def count_vowels(sentence):

    #declare a variable to store the vowels ,to handle case insensitivity for both uppercase and lowercase
    vowels = "aeiouAEIOU"

    #initialize the value of count to a zero
    count = 0

    #create a for loop that iterates through the sentence and also have an instance variable(char) to help you iterate
    for char in sentence:

      #create a branching statement that confirms if there are vowels inside the sentence (string)
        if char in vowels:

          #if the condition is true vowels are going to be counted as the loop iterates ,
          #this updates the count variable by one upon each iteration if condition is meet

            count += 1
    return count

#ask for user input and store it in the variable sentence
sentence=input("enter a sentence:")

#Call the function and specify the arguments ie(sentence),use a variable to store the value upon a return
result = count_vowels(sentence)

#display the total number of vowels in the sentence
print(f"Number of vowels in the sentence: {result}")

```

```python
# Challenge 2: Detecting Prime Numbers in a Range
# Description
# Given two integers start and end, print all primes between them (inclusive).
# Roles
# •	Coder: Implement is_prime(n) and loop with for n in range(start, end+1).
# •	Tester: Verify edge cases (start > end, ranges with no primes, small ranges).
# •	Presenter: Explain the nested loops and early exit via break.

#declare and define a  that accept parameters for the range of numbers to check if its prime
def print_primes_in_range(start, end):

#prints prime numbers within the users  range
  print(f"\nPrime numbers between {start} and {end+1} are")

  #outer for loop that increments the range values by one
  for num in range(start,end + 1):

       #checks if the instance variable is less than one
        if num > 1:

            for i in range(2, int(num**0.5) + 1):
                if (num % i) == 0:
                    break
            else:
                print(f"{num}")

# Example usage: print prime numbers between 20 and 50
start=int(input("enter start value"))
end=int(input("enter end value"))

print_primes_in_range(start,end)
```

```python
# Guess and Check with Limited Attempts
# Description
# Hard code a secret number between 1–50. Prompt the user up to 5 times to guess it. After each guess, print “Too low”, “Too high”, or congratulate and exit. If all attempts fail, reveal the secret.
# Roles
# •	Coder: Write a while loop controlling attempts and branching on guess comparison.
# •	Tester: Simulate correct and incorrect guess sequences; test boundary guesses.
# •	Presenter: Walk through loop flow and the else: clause on the while
secret_number=20
attempts=0

while attempts<5:
  guess=int(input("Guess a number:"))
  if guess < secret_number:
    print("Too low")
  elif guess > secret_number:
    print("Too high")
  elif guess ==secret_number:
    print("congartulations you made the correct guess")
    break
else:
    print(f"maximum attempts reached the secret number:{secret_number}")





```
