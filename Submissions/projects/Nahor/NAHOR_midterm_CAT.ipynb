```python
# Ask the user for the exam score
score = int(input("Enter exam score (0-100): "))

# criteria for grading

if 90 <= score <= 100:
    grade = "F"
elif 80 <= score <= 89:
    grade = "B"
elif 70 <= score <= 79:
    grade = "C"
elif 60 <= score <= 69:
    grade = "D"
elif 0 <= score < 60:
    grade = "F"
else:
    grade = None

# Print the result
if grade:
    print("grade :", grade)
else:
    print("Invalid input. Please enter a number between 0 and 100.")
```

```python
# Prompt the user to enter a sentence
sentence = input("Enter a sentence: ")

# Define the vowels
vowels = "aeiouAEIOU"

# Initialize a counter
count = 0

# Count vowels in the sentence
for char in sentence:
    if char in vowels:
        count += 1

# Print the total count
print("Total number of vowels:", count)
```

```python
#Create a function meeting the  requirements(positive no.)
def sum_digits():
    while True:
        try:
            num_str = input("Enter a positive integer: ")
            num = int(num_str)
            if num < 0:
                print("Please enter a positive integer.")
            else:
                break
        except ValueError:
            print("Invalid input. Please enter an integer.")

    total_sum = 0
    for digit in num_str:
        total_sum += int(digit)

    print(f"The sum of the digits of {num} is: {total_sum}")

# Use the function to execute the code
sum_digits()
```
