```python
#11.Grade Converter
#Asks user to input exam score an interger 0-100
score = int(input("Enter exam score (0-100): "))

#prints corresponding letter grade according to:
#90-100 A and limit to 100
if score > 100 or score < 0:
   print("Invalid input. Please enter a valid integer score between 0 and 100.")

elif score >= 90:
   print("Grade: A")

#80-89 B
elif score >= 80 :
  print("Grade: B")
#70-79 C
elif score >= 70:
  print("Grade: C")
#60-69 D
elif score >= 60:
  print("Grade: D")
#0-59 F
else:
  print("Grade: F")





```

```python
#12. Vowel Counter
#Asks user to enter a sentence
sentence = input("Enter a sentence: ") #prompts user
count = 0
#counts how many vowels case sensitive how many vowels appear in sentence

for char in sentence:
    if char in "aeiouAEIOU":
      count += 1
print("Number of vowels: ", count)
```

```python
#13.Decimal to Binary Converter
#Asks user to input a non-negative integer
user_input=int(input("Enter a non-negative integer: "))
binary = ""

#Prints its binary representation as a string of 0 and 1
while user_input > 0:
    binary = str(user_input % 2) + binary
    user_input //= 2
print("Binary representation: ", binary)
```

```python
#14 .Sum of Digits
#Asks user to input a positive integer

num_input=int(input("Enter a positive integer: "))
total = 0

#Calculates the sum of its digits
for digit in str(num_input):
    total += int(digit)
print("Sum of digits: ", total)
```
