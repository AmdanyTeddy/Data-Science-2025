```python
#code to evaluate user score and assign corresponding grades using if elif
#declare a function that accepts users score as arguments
def Grade_converter(s):

  #declare conditionals to handle differnt grades for different scores
        if s in range(90,101):
            return "A"
        elif s in range(80,90):
            return "B"
        elif s in range(70,80):
            return "C"
        elif s in range(60,70):
            return "D"
        elif s in range(0,60):
            return "F"
        else:
            return "Invalid"

#handle value error exception when user types in an invalid digit
try:
  #ask for users score
  score=int(input("Enter your score:"))
except ValueError:
  print("Invalid input")

#print results
print("Grade:",Grade_converter(score))


```

```python
#vowel counter
#declare a function to with the parameter sentence to reference the users input
def vowel_counter(sentence):

    #initialize the values of vowels in uppercase and lowercase to handle case insensitivty
    vowel='aeiouAEIOU'

    #initialize count to zero
    count=0

    #declare a for loop to iterate through the sentence
    for letter in sentence:

      #implement a conditional statement that checks if the iterable is in the sentence
      if letter in vowel:

        #increment the value of count
       count+=1
       #return count
    return count

#ask the user to enter a sentence
sentence=input("Enter a sentence:")
#print the total vowel count by passing sentence as an argument  in the function call
print("Number of vowels:",vowel_counter(sentence))
```

```python
#declare a binary converter function with a parameter to hold the passed arguments from the user
def binary_converter(nonneg):

 #initialize an empty binary string variable that will be appended by the loop
    binary=''

    #declare a while loop that checks if the number is a convertible integer
    while nonneg>0:

      #append the binary variable with results from the floor divison which evaluates to either a zero or a 1
        binary=str(nonneg%2)+binary
        nonneg=nonneg//2

        #return the binary value to be printed
    return binary

#ask for a non negative number from the user
nonneg=abs(int(input("Enter a non-negative integer:")))

#print the binary equivalent of the whole number by passing actual arguments obtained from user input to the function
print("Binary output:",binary_converter(nonneg))
```

```python
#function that adds digits from user number input
def addition(num):

  #initialize the sum variable to 0
  sum=0

  #iterate through digits in the num variable
  for digit in num:

    #update the sum after every iteration
    sum+=int(digit)

    #return results of total sum
  return sum

#ask for the user to input a positive number
num=input("Enter a positive number:")

#print the results from the addition function
print("Sum of digits:",addition(num))
```
