```python
#Assignment 	17
#Topic: Object-Oriented Programming
#Subtopics: Data Abstraction, Class Definition, Class Instances, Methods
#Challenge 1: ðŸ“¦ The Box Class
#Goal: Practice class creation, comparison between objects, and area/volume calculations.

#Challenge 1
#Write a class called Box that:
#Has attributes: length, width, height
#Has a method get_volume() that returns the volume (length Ã— width Ã— height)
#Has a method is_cube() that returns True if all sides are equal
#Has a method bigger(other) that returns the box with a larger volume

class Box:
    def __init__(self, length, width, height):
        self.length = length
        self.width = width
        self.height = height


    def get_volume(self): # Method get_volume() that returns the volume (length Ã— width Ã— height)
           return self.length * self.width * self.height

    def is_cube(self):# Method is_cube() that returns True if all sides are equal
            return self.length == self.width == self.height

    def bigger(self, other):# Method bigger(other) that returns the box with a larger volume
            return self if self.get_volume() > other.get_volume() else other
    def __str__(self):
            return f"Box with length {self.length}, width {self.width}, and height {self.height}"


#Calling the method to print output
box1 = Box(10, 10, 10)
box2 = Box(5, 5, 20)
print(box1.get_volume())
print(box2.is_cube())
print(box1.bigger(box2))




```

```python
#Challenge 2
#Write a class called Song that:
#Has attributes: title, artist, duration (in seconds)
#Has a method is_long() that returns True if the song is over 5 minutes (300 seconds)
#Has a method same_artist(other) that returns True if two songs have the same artist
#Has a method longer_song(other) that returns the song with the longer duration


class Song(object):
    def __init__(self, title, artist, duration):
        self.title = title
        self.artist = artist
        self.duration = duration

    def is_long(self):# Method is_long() that returns True if the song is over 5 minutes (300 seconds)
        return self.duration > 300


    def same_artist(self, other):# Method same_artist(other) that returns True if two songs have the same artist
        return self.artist == other.artist

    def longer_song(self, other):# Method longer_song(other) that returns the song with the longer duration
        return self if self.duration > other.duration else other

    def __str__(self):
        return f"{self.title} by {self.artist}"



song1 = Song("Bohemian Rhapsody", "Queen", 355)
song2 = Song("Imagine", "John Lennon", 210)
print(song1.is_long())
print(song1.same_artist(song2))
print(song1.longer_song(song2))



```
