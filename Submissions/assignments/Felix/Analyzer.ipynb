```python

# Final Project : Text Analyzer & Number Converter App
# This Individual Final Project  on Iteration each student to design and implement a “Text Analyzer & Number Converter” in Python, showcasing mastery of iteration concepts:
# for and while loops, nested loops, loop control (break/continue), Boolean flags, and sequence processing. Students will process user input strings, count and categorize characters,
# search for patterns, compute numeric statistics, and convert numbers into binary. The project is comprehensive—covering all major iteration patterns from Topics 1–3—and will be submitted via a GitHub classroom repository.

# Project Description
# Build a commandline Python application named analyzer.py that:
# String Analysis
# Counts: total characters, vowels, consonants, digits, and punctuation.
# Detects: the first and last occurrence of a usersupplied substring.
# Lists: all unique words (caseinsensitive), sorted alphabetically.
# Numeric Statistics
# Accepts a commaseparated list of integers from the user.
# Computes: sum, average (mean), minimum, and maximum using loops (no builtins like sum()).
# Flags: numbers divisible by 3 or 5, printing them with a label.
# Binary Converter
# Prompts the user for a positive or negative integer.
# Converts it to binary using the remainderdivision method (while loop), handling sign.
# Displays the binary string.
# Menu & Loop Control
# Presents a menu to the user to choose any of the above functions or exit.
# Uses a while True: loop with break/continue to manage the menu.

# Objectives for the Project
# Apply for loops to iterate over strings, lists, and ranges
# Use while loops for conditional repetition and number conversion.
# Employ break and continue to control loop flow.
# Implement nested loops for complex tasks (e.g., word extraction).
# Manage Boolean flags to detect conditions (e.g., substring found).
# Practice input/output and robust error handling.

# Requirements & Constraints
# No use of builtin aggregators: you must implement your own loops for counting and summing (no sum(), min(), max(), or count() methods).
# String processing: use both indexbased (range(len(s))) and direct (for ch in s) loops at least once each.
# Modular code: separate functionality into functions:
# analyze_text(s: str) -> dict
# stats_numbers(nums: list[int]) -> dict
# to_binary(n: int) -> str
# main_menu(): drives the program loop.
# Input validation: reject invalid inputs with error messages and reprompt.
# Documentation: include docstrings for every function and inline comments for loop logic.
# Style: adhere to PEP 8 style guidelines; use meaningful variable names.

# Deliverables
# analyzer.py: the project source file.
# README.md in the root of your GitHub folder:
# Brief project overview.
# Instructions for running (e.g., python analyzer.py).
# Sample input/output transcript.


# Programm that conducts String Analysis.
#1. Counts: total characters, vowels, consonants, digits, and punctuation.

def string_analysis(input_str):
    vowels = "aeiouAEIOU"
    consonants = "bcdfghjklmnpqrstvwxyzBCDFGHJKLMNPQRSTVWXYZ"
    punctuation = "!@#$%^&*()_+[]{}|;:,.<>?/"
    digits ="0123456789"

    total_chars =len(input_str)
    total_vowels = 0
    total_consonants = 0
    total_digits = 0
    total_punctuation = 0


    for char in input_str:
         if char in vowels:
                total_vowels += 1
         elif char.isalpha():
                total_consonants += 1
         elif char in digits:
             total_digits += 1
         elif char in punctuation:
            total_punctuation += 1

    print("\nString Analysis Results:")
    print(f"Total characters: {total_chars}")
    print(f"Total vowels: {total_vowels}")
    print(f"Total consonants: {total_consonants}")
    print(f"Total digits: {total_digits}")
    print(f"Total punctuation: {total_punctuation}")


#Detects: the first and last occurrence of a usersupplied substring.
def find_substring(input_str, substring):
    first_occurrence = input_str.find(substring)
    last_occurrence = input_str.rfind(substring)

    if first_occurrence == -1:
        print(f"\nThe substring  '{substring}': was not found in the string")
    else:
        print(f"'\nSubstring search Results:")

    print(f"First occurrence: {first_occurrence}")
    print(f"Last occurrence: {last_occurrence}")

# Lists: all unique words (caseinsensitive), sorted alphabeticall
def unique_words(input_str):
    words = input_str.split()#splits strings into words
    unique_words_set = set() #removes duplicates

    for word in words:
        unique_words_set.add(word.lower()) #makes it case insesitive

    unique_words_list = list(unique_words_set)
    unique_words_list.sort()# oders the words alphabetically

    return unique_words_list

# Accepts a commaseparated list of integers from the user.

def get_integer_list():
    input_str = input("Enter a list of integers separated by commas: ")
    try:
        integer_list = [int(num.strip()) for num in input_str.split(',')]
        print("Integer list:", integer_list)



 #Computes: sum, average (mean), minimum, and maximum using loops (no builtins like sum()).
        total=0
        minimum = integer_list[0]
        maximum = integer_list[0]

        for num in integer_list:
            total += num
            if num < minimum:
                minimum = num
            if num > maximum:
                maximum = num

        average = total / len(integer_list)
        print("Integer List Statistics (using loops):")

        print(f"Sum: {total}")
        print(f"Average: {average}")
        print(f"Minimum: {minimum}")
        print(f"Maximum: {maximum}")


 #Flags: numbers divisible by 3 or 5, printing them with a label.
        for num in integer_list:
          if num % 3 == 0 and num % 5 == 0:
            print(f"Divisible by both 3 and 5: {num}")
          elif num % 3 == 0:
            print(f"Divisible by 3: {num}")
          elif num % 5 == 0:
            print(f"Divisible by 5: {num}")

    except ValueError:
            print("Invalid input. Please enter a comma-separated list of integers.")


# Binary Converter
# Prompts the user for a positive or negative integer.
def binary_converter():
  try:
    num = int(input("Enter a positive or negative integer: "))
    if num >= 0:
      binary = bin(num)[2:]
      print(f"Binary representation of {num} is: {binary}")
    else:
      binary = bin(abs(num))[3:]
      print(f"Binary representation of {num} is: -{binary}")
  except ValueError:
    print("Invalid input. Please enter a valid integer.")


# Menu & Loop Control
# Presents a menu to the user to choose any of the above functions or exit.
# Uses a while True: loop with break/continue to manage the menu.
def main_menu():
        print("\nMenu:")
        print("1. String Analysis")
        print("2. Substring Search")
        print("3. Unique Words")
        print("4. Integer List Statistics")
        print("5. Binary Converter")
        print("6. Exit")

while True:
        main_menu()
        choice = input("Enter your choice (1-6): ")

        if choice == '1':
            user_input = input("Enter a string: ")
            string_analysis(user_input)

        elif choice == '2':
            user_input = input("Enter a string: ")
            substring_to_find = input("Enter a substring to find: ")
            find_substring(user_input, substring_to_find)

        elif choice == '3':
            user_input = input("Enter a string: ")
            unique_word_list = unique_words(user_input)

            print("\nUnique words in the string:")
            for word in unique_word_list:
                print(word)

        elif choice == '4':
            get_integer_list()

        elif choice == '5':
            binary_converter()

        elif choice == '6':
            print("Exiting the program. Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a number between 1 and 6.")


        input("Press Enter to continue...")

        continue




















```
