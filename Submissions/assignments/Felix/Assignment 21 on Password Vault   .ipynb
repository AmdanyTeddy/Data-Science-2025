```python
from logging import Manager
import json
import base64
import os

VAULT_FILE = "vault . json"

def load_vault():
  """Load vault from file"""
  if not os.path.exists(VAULT_FILE):
    return{}
  with open (VAULT_FILE, "r")as f :
    return json.load(f)


def save_vault(vault):
  """Save password vault to file."""
  with open (VAULT_FILE, "w") as f:
     json.dump(vault, f, indent=4)

def encrypt_password(password):
  """Encrypt password using base64"""
  return base64.b64encode(password.encode("utf-8")).decode("utf-8")


def decrypt_password(encrypted_password):
  """Decrypt password using base64"""
  return base64.b64decode(encrypted_password.encode("utf-8")).decode("utf-8")


def add_entry():
  """Add a new password entry."""
  name = input("Enter  entry name (e.g ,Gmail,Facebook):")
  username = input("Enter username/email:")
  password = input("Enter password:")

  vault = load_vault()
  vault[name] = {
      "username": username,
      "password":
      encrypt_password(password)

  }
  save_vault(vault)
  print(f"[+] Entry '{name}' added successfully!")


def retrieve_entry():
  """Retrieve a password entry"""
  keyword = input("Enter service name to retrive: ")

  vault = load_vault()

  if keyword in vault:
    username = vault[keyword]["username"]
    password = decrypt_password(vault[keyword]["password"])
    print(f"/nService: {keyword}\nUsername:{username}\nPassword:{password}")
  else:
       print("[-] Entry not found.")

def delete_entry():
  """Delete a password entry."""
  keyword = input("Enter service name to delete: ")
  vault = load_vault()

  if keyword in vault:
    del vault[keyword]
    save_vault(vault)
    print(f"[+] Entry '{keyword}'deleted successfully.")
  else:
    print("[-]Entry not found.")


def main():
  while True:
    print("\n ---Password Manager---")
    print("1.Add new entry")
    print("2.Retrieve password")
    print("3.Delete entry")
    print("4. Exit")

    choice = input("choose an option(1-4): ")


    if choice == "1":
       add_entry()
    elif choice == "2":
       retrieve_entry()
    elif choice == "3":
      delete_entry()
    elif choice == "4":
      print("Exiting password Manager")

      break
    else:
      ("Invalid choice .Try again.")


if __name__=="__main__":
  main()
```
