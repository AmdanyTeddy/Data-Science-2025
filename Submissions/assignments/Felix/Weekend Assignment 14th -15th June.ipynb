```python
#Problem 1:
#Write a Python program to find the cube root of any float between -1000 and 1000 using bisection search. Handle negative cubes correctly.
#Problem 2:
#Find the square root of any number between 0 and 1 using Newton-Raphson.
#Problem 3 (From Topic 5 - Approximation):
#Modify your square root program to use the approximation method with incremental steps (like 0.0001) and compare how many steps it takes vs. bisection search for the same value of x = 12345.
#Problem 4 (Guessing game):
#Create a guessing game where the computer randomly chooses a number and you guess it using bisection search. The program should guide you: "Too high", "Too low", or "Correct".


#Problem 1
#Python program that finds the cube root of any float between -1000 and 1000 using bisection search

def find_cube_root(number , epsilon = 0.001):
    if number >= 0:
        low = 0
        high = max(1.0 , number)
    else:
        low = min(-1.0 , number)
        high = 0
    guess = (low + high) / 2.0
  #bisection serarch loop
    while abs(guess**3 - number) >= epsilon:
        if guess**3 < number:
            low = guess
        else:
            high = guess
        guess = (low + high) / 2.0
    return guess

num= float(input("Enter a number between -1000 and 1000: "))

if -1000 <= num <= 1000:
    print("The cube root of", num, "is", find_cube_root(num))
else:
    print("The number is out of range.")
```

```python
#Problem 2
#Find the square root of any number between 0 and 1 using Newton-Raphson.
def newton_raphson_sqrt(number , epsilon = 0.00001):
    if number < 0 or number >= 1:
        return "Invalid input. Number must be between 0 and 1."

    guess = number / 2.0
    while abs(guess**2 - number) >= epsilon:
        guess = guess -((guess**2 - number) / (2 * guess))
    return guess

num = float(input("Enter a number between 0 and 1: "))

result = newton_raphson_sqrt(num)
print("The square root of", num, "is approximately", result)
```
