```python
#Final Assignment
#ðŸš€ Assignment Task 1:
#Create a function that:
#Asks the user for a word.
#Returns only the consonants from that word.

def keep_consonants(word):
    vowels = "aeiou"
    result = ""
    for char in word:
        if char not in vowels:
            result += char
    return result

print(keep_consonants("dictionary"))  # Output: bbs

```
