```python
#Instructions to create the program
#1.prompt the user to input a temperature in celcius using the input function

temperature_celcius = input("Enter the temperature in celcius: ")
print("The temperature in celcius is:", temperature_celcius)
#2. converts the input string to float
temperature_celcius = float(temperature_celcius)
print("The temperature in celcius is:", temperature_celcius)

#3.calculate the equivalent Fahrenheit temperature using formula(fahrenheit = (celsius * 9/5 + 32)
fahrenheit = (temperature_celcius * 9/5) + 32

print("The equivalent Fahrenheit temperature is:", fahrenheit)

#5 . Use conditional statements to display messages based on the Celsius temperature

# below 0: "its freezing cold!"
if temperature_celcius < 0:
    print("It's freezing cold!")

# 0 to 20 :"it's a bit chilly."
elif temperature_celcius >= 0 and temperature_celcius <= 20:
    print("It's a bit chilly.")

# 21 to 30 :"nice and warm."
elif temperature_celcius >= 21 and temperature_celcius <= 30:
    print("Nice and warm.")

# above 30: "it's hot!"
else:
    print("It's  quite hot!")



```
