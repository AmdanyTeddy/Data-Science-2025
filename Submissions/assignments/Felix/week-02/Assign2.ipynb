```python
#calculations of  volume and surface area of a cylinder
# Mesurements of cylinder

radius = 3.0 # radius of cylinder (in units)
height = 5  # height of cylinder (in units)
pi = 3.14159 #Remains constant in caculations

#Volume calculation
volume = pi * radius * radius  * height #Volume is obtained by getting the base area of the cylinder multiply by height

#Surface_Area Calculation
surface_area = 2 * pi * (radius * radius) + 2 * (pi * radius * height) # obtained by summing up the area of both circular bases and lateral surface

print("volume:" , volume  )
print( "surface_area:", surface_area)
```
