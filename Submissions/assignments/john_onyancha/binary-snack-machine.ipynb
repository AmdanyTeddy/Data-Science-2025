```python
# summary
# guess and check -try every posibilty
# perfect square-if guess**2==2,then x is a perfect square
# binary conversion-divide by 2 collect remainders,reverse them
# float inaccuracy-some decimal like 0.1 cant be exactly represented in binary

def float_to_binary(num):
    # Split the number into integer and fractional parts
    intp = int(num)
    decp = num - intp

    # Convert integer part to binary
    if intp == 0:
        int_binary = "0"
    else:
        int_binary = ""
        while intp > 0:
            int_binary = str(intp % 2) + int_binary
            intp //= 2

    # Convert fractional part to binary
    dec_binary = ""
    count = 0
    while decp > 0 and count < 8:
        decp *= 2
        bit = int(decp)
        dec_binary += str(bit)
        decp -= bit
        count += 1

    # Combine both parts
    if decp:
        return f"{int_binary}.{dec_binary}"
    else:
        return int_binary

#ask for the user input
print(" Welcome to the Binary Snack Machine!")
price = float(input("Enter snack price (between 0 and 10): "))


if price>=0 and price<=10.00000000:
    binary_price = float_to_binary(price)
    print(f"Binary price: {binary_price}")
    print(" Enjoy your snack! ")
else:
    print("Please enter a valid price between 0 and 10.")
```
