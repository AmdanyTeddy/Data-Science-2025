```python
#counting unique common characters in two strings

s="may the fourth$ ...."
s1="revenge of the sixth"
common=set(s).intersection(set(s1))
print(f"common characters:{common}")

unicommon=[]
count=0
letters=("abcdefghijklmnopqrstuvwxyz ")
for i in common:
  if i in letters:
    unicommon.append(i)
    count+=1
print(f" unique common characters:{unicommon}")
print(f"number of unique common characters:{count}")
```
