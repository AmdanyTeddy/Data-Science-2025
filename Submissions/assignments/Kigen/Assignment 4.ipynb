```python
#5).BRANCHING:CONDITIONAL STATEMENTS
#Branching allows a program to choose between different paths of execution based on conditions.
#In Python, we use conditional statements to control the flow of the program.)
#a)if statement
#Executes a block of code only if a specified condition is True.
x = 5
if x > 0:
    print("x is positive")

```

```python
#b).elif Statement in Python
#The elif keyword stands for "else if". It's used to check multiple conditions after an initial if statement.
#It allows your program to respond differently depending on various conditions.
x = -10

if  x > 0:
    print("x is positive")
elif x == 0:
    print("x is zero")
else:
 print("x is negative")
```
