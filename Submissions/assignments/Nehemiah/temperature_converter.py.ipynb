```python
# prompts the user input
celcius = float(input("Enter the temperature (degrees c) : "))

# calculate the equivalent Fahrenheit temp
Fahrenheit = (celcius * 9/5) + 32

# uses conditional statements if, elif and else
if celcius < 0:
  print("It's freezing cold")

elif celcius <= 20:
  print("Its a bit chilly")

elif celcius <= 30:
  print("Its nice and warm")

else:
  print("Its quite hot!")

print(f"temperaturein celcius : {celcius}")
print(f"temperature in fahrenheit : {Fahrenheit}")
```
