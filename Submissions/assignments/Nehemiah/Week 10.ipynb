1.	Write a program that:
•	Takes a list of names
•	Removes duplicates
•	Sorts alphabetically

```python
# create a list of names
names = ["Dan", "Alison", "Colloh", "Kim", "Purry", "Alison", "Jack", "Daniel", "Kigen", "Kim"]

# convert list to a set
unique_names = list(set(names))

unique_names.sort()

print(names)
print("Sorted list", unique_names)
```
