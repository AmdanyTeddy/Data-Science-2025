```python
# Challenge: Binary Snack Machine (15â€“20 min, Individual or Pairs)
# Write a Python program that asks the user for a price (a float between 0 and 10), then prints its binary representation, using guess-and-check logic for the fractional part.

print("Welcome to the binary Snack Machine")

# Initiate the try except to capture invalid values
while True:
  try:

    price = input("Enter snack price (between 0 and 10): ")
    num = float(price)
    if 0 <= num <=10:
      print("Bon appetite! Enjoy your snack.")
      break
    else:
      print("Price must be between 0 and 10.")
  except ValueError:
    print("Invalid input! Please enter your price.")

# separate integer and fractional parts
integer = int(num)
fractional = num - integer

# converts the integer part to binary(manually converts without using built-in functions)
def convert_to_binary(integer):
  if integer == 0:
    return "0"

  binary_digits = []
  while integer > 0:
    remainder = integer % 2
    binary_digits.insert(0, str(remainder))
    integer = integer // 2

  return "".join(binary_digits)



```
