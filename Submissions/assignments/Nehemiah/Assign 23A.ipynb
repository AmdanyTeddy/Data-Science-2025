**a)Checking if a list has duplicates**

**Complexity: Linear time complexity O(n)**

For exaample:
Imagine you have an empty guest book (a set) where everyone signs as they arrive.

"Alice" arrives. You check the guest book. Her name isn't in it, so she signs.

"Bob" arrives. His name isn't in the book, so he signs.

"Carol" arrives. Her name isn't in the book, so she signs.

Another "Alice" arrives. You check the guest book. Her name is already there! You found a duplicate!

You only had to do one quick check (is the name in the book?) for each person. For n people, that's n checks. This is O(n).

**b) Sorting a dataset of 1M entries**

**Complexity : Time complexity O(n log n)**

For example:
Imagine you have to sort 1,000,000 library books by title. You can't compare them all individually—that would be O(n²) and take forever.

Instead, you use a smart strategy like **Mergesort**:

**Divide**: Split the massive pile of 1M books into 2 piles of 500,000. Split those into 4 piles of 250,000. Keep splitting until you have 1,000,000 piles of just 1 book each. A pile of 1 book is already sorted!

**Conquer (Merge)**: Now, start merging small sorted piles back together. Merge two 1-book piles into a sorted 2-book pile. Merge two 2-book piles into a sorted 4-book pile. This merging process is very efficient.

The number of times you have to split the pile is log₂(1,000,000) ≈ 20. For each of those 20 "levels", you have to look at all n (1,000,000) books during the merge step.

So the total work is n * log(n) = 1,000,000 * 20, which is much, much better than a naive n² = 1,000,000,000,000 (1 trillion) comparisons

**c) Training k-NN on n samples**

Complexity : Constant time complexity O(1) or Linear time complexity O(n)

For example:
 Imagine you are an art expert learning to identify painters. Your friend is teaching you k-NN style.

"Training": Your friend simply shows you 1,000 paintings and tells you who painted each one. You don't try to learn rules like "Van Gogh uses swirly brushes." You just memorize the entire collection. This is the "training" phase. The effort is proportional to how many paintings you see—it's O(n) to store them all in your memory. It's very fast per painting, so we often say this step is trivial (O(1) per sample).

"Prediction": The hard work comes later. When you see a new, unknown painting, you have to compare it to every single one of the 1,000 paintings in your memory to find the 5 most similar ones (k=5). This is very slow (O(n) for each new painting) because you have to do 1,000 comparisons.

**d) Running BFS on a graph with V vertices, E edges**

Complexity : **Linear Time complexity** represented by O(V + E) for algorithms that operate on graphs, where V is the number of vertices and E is the number of edges

2.Write your own binary search and prove its Θ(log n).

fib_recur(n) vs an iterative Fibonacci implementation.
Plot time taken for n = 10…40.


```python
def binary_search(arr, target):
  low = 0
  high = len(arr) - 1

  while low <= high:
    mid = (low+high) // 2
    guess = arr[mid]

    if guess == target:
      return mid
    elif guess < target:
      low = mid + 1
    else:
      high = mid - 1
  return -1

sorted_list = [11, 13, 15, 17, 19, 21, 23]
print(binary_search(sorted_list, 9))
print(binary_search(sorted_list, 11))
```

```python
# Runtime analysis
# uses time module

import time
import matplotlib.pyplot as plt

# recursive fibonacci
def fib_recur(n):
  if n <= 1:
    return n
  else:
    return fib_recur(n-1) + fib_recur(n-2)

# iterative fibonacci
def fib_iter(n):
  a, b = 0, 1
  for _ in range(n):
    a, b = b, a + b
  return a

n_values = range(10, 41)
recur_times = []
iter_times = []

# timing
for n in n_values:
  start = time.time()
  fib_recur(n)
  end = time.time()
  recur_times.append(end - start)

  start = time.time()
  fib_iter(n)
  end = time.time()
  iter_times.append(end - start)

# plotting
plt.plot(n_values, recur_times, label="Recursive", marker="o")
plt.plot(n_values, iter_times, label="Iterative", marker="o")
plt.xlabel("n")
plt.ylabel("Time(seconds)")
plt.title("Fibonacci Runtime: Recursive Vs Iterative")
plt.legend()
plt.grid(True)
plt.show()
```
