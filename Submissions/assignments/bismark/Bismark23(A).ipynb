1.Classify complexity of these tasks:
     -Checking if a list has duplicates.
     -Sorting a dataset of 1M entries.
     -Training k-NN on n samples.
     -Running BFS on a graph with V vertices, E edges.



A simple way to classify the complexity of these tasks is to think about how much "work" a computer has to do as the amount of data gets bigger.

Checking for duplicates: If you check every item against every other item, the work grows really fast, like "N-squared" (

N
2

), where N is the number of items. A more clever way using a "hash set" is much faster, only taking a time equal to the number of items, or just

N
.

Sorting a million items: The most common way to sort data is much more efficient than the "N-squared" method. It’s typically "N log N" (

NlogN
), which is good enough for huge datasets.

Training k-NN: This one is easy to train because the computer just needs to memorize all the data points. The hard work comes later when you ask it to find the nearest neighbors for a new point.

Running BFS: To find a path in a graph, BFS is very efficient. It checks every connection and every point in the graph just once. The time it takes is proportional to the number of vertices and edges, or

V+E

Plot time taken for n = 10…40.



```python
import time
import matplotlib.pyplot as plt

def binary_search(arr, target):
    """
    Performs a binary search on a sorted array.

    Args:
        arr (list): A sorted list of elements.
        target: The value to search for.

    Returns:
        int: The index of the target if found, otherwise -1.
    """
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        # Check if the target is at the middle
        if arr[mid] == target:
            return mid
        # If target is greater, search the right half
        elif arr[mid] < target:
            low = mid + 1
        # If target is smaller, search the left half
        else:
            high = mid - 1

    return -1

def fib_recur(n):
    """
    Calculates the nth Fibonacci number using recursion.
    WARNING: This is extremely inefficient for larger n due to repeated calculations.
    """
    if n <= 1:
        return n
    return fib_recur(n - 1) + fib_recur(n - 2)

def fib_iter(n):
    """
    Calculates the nth Fibonacci number using an iterative approach.
    This is much more efficient than the recursive method.
    """
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a

def plot_fibonacci_runtimes():
    """
    Measures and plots the runtime of recursive vs. iterative Fibonacci.
    """
    # The range is limited to prevent excessively long runtimes for the recursive function.
    n_values = list(range(10, 41))
    recursive_times = []
    iterative_times = []

    print("Measuring runtimes...")
    for n in n_values:
        print(f"n = {n}...")

        # Measure recursive time
        start_time_rec = time.time()
        fib_recur(n)
        end_time_rec = time.time()
        recursive_times.append(end_time_rec - start_time_rec)

        # Measure iterative time
        start_time_iter = time.time()
        fib_iter(n)
        end_time_iter = time.time()
        iterative_times.append(end_time_iter - start_time_iter)

    # Plotting the results
    plt.figure(figsize=(10, 6))
    plt.plot(n_values, recursive_times, label='Recursive (O(φ^n))', color='red')
    plt.plot(n_values, iterative_times, label='Iterative (O(n))', color='blue')

    plt.xlabel('n (The nth Fibonacci number)')
    plt.ylabel('Time (seconds)')
    plt.title('Recursive vs. Iterative Fibonacci Runtime')
    plt.legend()
    plt.grid(True)
    plt.show()

# Run the plotting function
if __name__ == '__main__':
    plot_fibonacci_runtimes()
```
