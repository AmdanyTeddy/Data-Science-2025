```python
# Final Project : Text Analyzer & Number Converter App
# This Individual Final Project  on Iteration each student to design and implement a “Text Analyzer & Number Converter” in Python, showcasing mastery of iteration concepts: for and while loops, nested loops, loop control (break/continue), Boolean flags, and sequence processing. Students will process user input strings, count and categorize characters, search for patterns, compute numeric statistics, and convert numbers into binary. The project is comprehensive—covering all major iteration patterns from Topics 1–3—and will be submitted via a GitHub classroom repository.

# Project Description
# Build a command line Python application named analyzer.py that:
# 1.	String Analysis
text = input("Enter the text to analyze:\n")
substring = input("Enter a substring to search for: ")

import string

def analyze_text(text):
#•	Lists: all unique words (case insensitive), sorted alphabetically.
    vowels = 'aeiouAEIOU'
    punctuation = string.punctuation
    digits = '0123456789'
#•	Counts: total characters, vowels, consonants, digits, and punctuation.
    total_chars = 0
    vowel_count = 0
    consonant_count = 0
    digit_count = 0
    punctuation_count = 0

    for ch in text:
        total_chars += 1

        if ch in vowels:
            vowel_count += 1
        elif ch.isalpha():
            consonant_count += 1
        elif ch in digits:
            digit_count += 1
        elif ch in punctuation:
            punctuation_count += 1

    print("\n--- Character Analysis ---")
    print(f"Total characters: {total_chars}")
    print(f"Vowels: {vowel_count}")
    print(f"Consonants: {consonant_count}")
    print(f"Digits: {digit_count}")
    print(f"Punctuation: {punctuation_count}")
#•	Detects: the first and last occurrence of a user supplied substring.
def find_substring_occurrences(text, substring):
    first = -1
    last = -1
    text_length = len(text)
    sub_length = len(substring)

    for i in range(text_length - sub_length + 1):
        match = True
        for j in range(sub_length):
            if text[i + j] != substring[j]:
                match = False
                break
        if match:
            if first == -1:
                first = i
            last = i

    print("\n--= Substring Occurrence =--")
    if first == -1:
        print(f"Substring '{substring}' not found.")
    else:
        print(f"First occurrence of '{substring}': index {first}")
        print(f"Last occurrence of '{substring}': index {last}")

def list_unique_words(text):
    words = []
    word = ''
    for ch in text:
        if ch.isalnum():
            word += ch.lower()
        elif word:
            words.append(word)
            word = ''
    if word:
        words.append(word)

    unique_words = []
    for w in words:
        exists = False
        for uw in unique_words:
            if w == uw:
                exists = True
                break
        if not exists:
            unique_words.append(w)

    for i in range(len(unique_words)):
        for j in range(i + 1, len(unique_words)):
            if unique_words[i] > unique_words[j]:
                unique_words[i], unique_words[j] = unique_words[j], unique_words[i]

    print("\n--- Unique Words (sorted, case-insensitive) ---")
    for w in unique_words:
        print(w)

def main():
    print("=== Text Analyzer ===")
    analyze_text(text)
    find_substring_occurrences(text, substring)
    list_unique_words(text)

if __name__ == "__main__":
    main()




```

```python
# 2.	Numeric Statistics
def numeric_statistics():
    while True:
        try:
            user_input = input("Enter a list of integers separated by commas: ")

# •	Accepts a comma separated list of integers from the user.
            str_numbers = user_input.split(",")
            numbers = []
            for num in str_numbers:
                num = num.strip()
                if num == "":
                    continue
                numbers.append(int(num))

            if not numbers:
                print("No valid numbers entered. Try again.")
                continue

# •	Computes: sum, average (mean), minimum, and maximum using loops (no built ins like sum()).
            total = 0
            count = 0
            min_num = numbers[0]
            max_num = numbers[0]

            print("\nNumbers divisible by 3 or 5:")
            for number in numbers:
                total += number
                count += 1


                if number < min_num:
                    min_num = number
                if number > max_num:
                    max_num = number


                if number % 3 == 0 or number % 5 == 0:
                    print(f"{number} is divisible by 3 or 5")

            average = total / count

# •	Flags: numbers divisible by 3 or 5, printing them with a label.
            print("\nStatistics Summary:")
            print("Sum:", total)
            print("Average:", average)
            print("Minimum:", min_num)
            print("Maximum:", max_num)

            break

        except ValueError:
            print("Invalid input. Please enter integers separated by commas.")
            continue

numeric_statistics()
```

```python
# 3.	Binary Converter
def binary_converter():
    while True:
        try:
# •	Prompts the user for a positive or negative integer.
            user_input = input("Enter an integer (positive or negative): ")


            number = int(user_input)

            if number == 0:
                print("Binary: 0")
                break
# •	Converts it to binary using the remainder division method (while loop), handling sign.
            is_negative = number < 0
            number = abs(number)

            binary_str = ""
            while number > 0:
                remainder = number % 2
                binary_str = str(remainder) + binary_str
                number = number // 2

            if is_negative:
                binary_str = "-" + binary_str
# •	Displays the binary string.
            print("Binary:", binary_str)
            break
        except ValueError:
            print("Invalid input. Please enter a valid integer.")
            continue


binary_converter()

```
