```python
# Fixed secret number
secret = 4

# Allow up to 10 guesses
for attempt in range(1, 11):  # attempt goes from 1 to 10
    try:
        guess = int(input(f"Attempt {attempt}/10 - Guess the secret number (1–10): "))
        if guess == secret:
            print(f"Hooray! The secret number is {secret}")
            break
        else:
            print("Wrong guess. Try again!")
    except ValueError:
        print("Please enter a valid number!")

# If the loop finishes without a correct guess
else:
    print(f"Sorry, you've used all 10 attempts. The secret number was {secret}.")
```
