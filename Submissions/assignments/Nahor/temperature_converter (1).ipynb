```python
#Prompt the user to input a temperature in celsius.
temp_incelsius = float(input("Enter temperature in celsius: "))
# Convert the the temperature to Fahreheit.
temp_infaren = (temp_incelsius * 9/5) + 32
#Display the original temp in celsius and in fahrenheit.
print(f"Celsius Temperature: {temp_incelsius}")
print(f"Fahrenheit Temperature: {temp_infaren}")
#Set statements to be displayed based on the temperature provided.
if temp_incelsius < 0 :
   print("It,s freezing cold!")
elif temp_incelsius <= 20 :
   print("It,s abit chilly.")
elif temp_incelsius <= 30 :
    print("Nice and warm.")
else :
    print("It,s quite hot!")



```
