```python
# Challenge 1: Digit Frequency Counter
# Goal: Practice for loops, string handling, and dictionaries to count how often each digit appears.
# Description
# 1.Prompt the user:
num_str = input("Enter a positive integer: ")
# 2.	Validate input is all digits (str.isdigit())
if not num_str.isdigit():
    print("Invalid input. Please enter a positive integer made up of digits only.")
else:
# 3. Initialize a dictionary
     counts = {}
# 4. Iterate over each character in the string (for ch in num_str:) :
# •	Convert ch to an integer if desired, or keep as string key.
     for ch in num_str:
       counts[ch] = counts.get(ch, 0) + 1

# 5.	After the loop, print each digit and its frequency in ascending order:
     for digit in sorted(counts):
         print(f"Digit {digit}: {counts[digit]} time(s)")
# 6.	Example Run:
# css
# CopyEdit
# Enter a positive integer: 12234331
# Digit 1: 3 time(s)
# Digit 2: 2 time(s)
# Digit 3: 3 time(s)
# Digit 4: 1 time(s)





```

```python
# Challenge 2: Guess the Number Game with Limited Attempts
# Goal: Combine while loops, branching, and input/output to create an interactive game.
# Description
# 1.Import the built in random module.
# 2.	Generate a random integer between 1 and 50:
import random


secret = random.randint(1, 50)
# 3.	Set a counter attempts = 0 and a maximum max_attempts = 5.
attempts = 0
max_attempts = 5
# 4.	Use a while loop (while attempts < max_attempts:) to allow up to 5 guesses:
while attempts < max_attempts:
    try:
        # Prompt the user
        guess = int(input(f"Attempt {attempts + 1}/{max_attempts}: Your guess? "))
        attempts += 1

        # Branching logic
        if guess == secret:
            print(f"🎉 Correct! You guessed the number in {attempts} tries.")
            break
        elif guess < secret:
            print("Too low! Try again.")
        else:
            print("Too high! Try again.")
    except ValueError:
        print("Please enter a valid integer.")
# print(" Correct! You guessed the number in", attempts, "tries.")
# break
# •	Elif guess < secret:

# print("Too low! Try again.")
# •	Else:

# print("Too high! Try again.")
# 5.	After the loop, if the user didn’t guess correctly:

# else:
#     print(" Out of attempts. The number was", secret)
# 6.	Example Run:
# Attempt 1/5: Your guess? 25
# Too low! Try again.
# ...
#  Correct! You guessed the number in 4 tries.
else:
    print(f"Out of attempts. The number was {secret}.")


# References
# 1.	Python string methods: str.isdigit()
# 2.	For loop syntax: for x in iterable:
# 3.	While loop overview: “Python’s while loop repeats as long as condition is true”
# 4.	Branching examples in loops: GeeksforGeeks
# 5.	Dictionary .get() for counting
# 6.	random.randint() documentation
# 7.	Sorting dictionary keys: Python sorted() built-in
# 8.	F strings for output formatting
# 9.	Break and else clause on loops
# 10.	Input conversion: int(input())
```
