```python
def cube_root_bisection(x, epsilon=0.0001):
    low = min(-1.0, x)
    high = max(1.0, x)
    guess = (high + low) / 2.0

    while abs(guess**3 - x) >= epsilon:
        if guess**3 < x:
            low = guess
        else:
            high = guess
        guess = (high + low) / 2.0

    return guess

# Example usage:
x = float(input("Enter a float between -1000 and 1000: "))
print(f"Cube root of {x} is approximately {cube_root_bisection(x)}")
```

```python
def sqrt_newton_raphson(x, epsilon=0.000001):
    guess = x / 2.0
    while abs(guess*guess - x) >= epsilon:
        guess = guess - ((guess**2 - x) / (2 * guess))
    return guess

# Example usage:
x = float(input("Enter a number between 0 and 1: "))
if 0 < x < 1:
    print(f"Square root of {x} is approximately {sqrt_newton_raphson(x)}")
else:
    print("Input must be between 0 and 1.")
```

```python
def sqrt_approx(x, epsilon=0.0001):
    step = epsilon**2
    num_guesses = 0
    guess = 0.0

    while abs(guess**2 - x) >= epsilon and guess <= x:
        guess += step
        num_guesses += 1

    return guess, num_guesses

def sqrt_bisection(x, epsilon=0.0001):
    low = 0.0
    high = x
    guess = (high + low) / 2.0
    num_guesses = 0

    while abs(guess**2 - x) >= epsilon:
        if guess**2 < x:
            low = guess
        else:
            high = guess
        guess = (high + low) / 2.0
        num_guesses += 1

    return guess, num_guesses

x = 12345
approx_result, approx_steps = sqrt_approx(x)
bisection_result, bisection_steps = sqrt_bisection(x)

print(f"Approximation sqrt({x}): {approx_result} in {approx_steps} steps")
print(f"Bisection sqrt({x}): {bisection_result} in {bisection_steps} steps")
```
