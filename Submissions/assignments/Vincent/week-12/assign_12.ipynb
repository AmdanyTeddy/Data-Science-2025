```python
def check_age(age):
    assert age >= 0, "Age cannot be negative"
    print("Age is okay.")

# ðŸ§ª Example:
check_age(25)   # Age is okay.
check_age(-3)   # AssertionError: Age cannot be negative
```

```python
def is_even(n):
    if not isinstance(n, int):
        raise Exception("Input must be an integer")
    return n % 2 == 0

# ðŸ§ª Example:
print(is_even(10))   # True
print(is_even("a"))  # Exception: Input must be an integer
```
