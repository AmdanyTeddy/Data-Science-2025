```python
#cloning - creating a new list with the same content
original =["banana", "apple",  "mango"]
copy1 =original[:]
copy2 = list(original)
copy3 = original.copy()
copy1.append("orange")
print("original;", original)
print("copy1;", copy1)
```

```python
# aliasing
colors = ['red','green','blue']
alias_colors = colors
alias_colors.append('yellow')
print('original list', colors)
print('alias list', alias_colors)
```

```python
import copy
nested = [[1, 2][3, 4]]
shallow_copy = copy.copy(nested)
deep_copy = copy.deepcopy(nested)
shallow_copy[0][0] = 5
```

```python
#square numbers using a list comprehension
nums = [1, 2, 3, 4, 5]
squares = [n**3 for n in nums]
print('Original;', nums)
print('squares;', squares)
```
