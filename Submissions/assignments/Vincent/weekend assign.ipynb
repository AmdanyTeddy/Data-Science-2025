```python
def cube_root(x):
    epsilon = 0.0001
    low = min(-1.0, x)
    high = max(1.0, x)
    guess = (high + low) / 2.0

    while abs(guess ** 3 - x) >= epsilon:
        if guess ** 3 < x:
            low = guess
        else:
            high = guess
        guess = (high + low) / 2.0

    return guess

# Example
x = float(input("Enter a float between -1000 and 1000: "))
print(f"The cube root of {x} is approximately {cube_root(x)}")
```

```python
def newton_raphson_sqrt(x):
    epsilon = 0.000001
    guess = x / 2.0
    while abs(guess ** 2 - x) >= epsilon:
        guess = guess - ((guess ** 2 - x) / (2 * guess))
    return guess

# Example
x = float(input("Enter a float between 0 and 1: "))
if 0 < x < 1:
    print(f"The square root of {x} is approximately {newton_raphson_sqrt(x)}")
else:
    print("Input must be between 0 and 1.")
```

```python
def sqrt_incremental(x, step=0.0001):
    guess = 0.0
    num_guesses = 0
    while guess**2 < x:
        guess += step
        num_guesses += 1
    return guess, num_guesses

def sqrt_bisection(x):
    epsilon = 0.0001
    low = 0
    high = x
    guess = (high + low) / 2.0
    num_guesses = 0
    while abs(guess ** 2 - x) >= epsilon:
        if guess ** 2 < x:
            low = guess
        else:
            high = guess
        guess = (high + low) / 2.0
        num_guesses += 1
    return guess, num_guesses

x = 12345
inc_result, inc_steps = sqrt_incremental(x)
bis_result, bis_steps = sqrt_bisection(x)

print(f"Incremental sqrt of {x}: {inc_result} in {inc_steps} steps")
print(f"Bisection sqrt of {x}: {bis_result} in {bis_steps} steps")
```
