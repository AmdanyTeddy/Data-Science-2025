```python
def hanoi_moves(n, source, target, spare):
    """
    Solves the Towers of Hanoi problem recursively.
    Returns a list of move instructions.
    """
    if n == 1:
        return [f"Move from {source} to {target}"]
    else:
        moves = hanoi_moves(n - 1, source, spare, target)
        moves.append(f"Move from {source} to {target}")
        moves += hanoi_moves(n - 1, spare, target, source)
        return moves

# ðŸ§ª Examples:
print(hanoi_moves(2, 'A', 'C', 'B'))
# âžœ ['Move from A to B', 'Move from A to C', 'Move from B to C']

print(hanoi_moves(3, 'A', 'C', 'B'))
# âžœ ['Move from A to C', 'Move from A to B', 'Move from C to B', 'Move from A to C', 'Move from B to A', 'Move from B to C', 'Move from A to C']
```
