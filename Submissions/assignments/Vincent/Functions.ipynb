```python
#writing and using functions
def is_even(i):
  """Returns Trye if the input is even, otherwise returns False."""
  if i % 2 == 0:
    return True
  else:
    return False
```

```python
def eval_quadratic(a, b, c, x):
  return a * x ** 2 + b * x + c
print(eval_quadratic(1, 1, 1, 1))
```

```python
def sum_odd(a,b):
  total = 0
  for i in range(a, b + 1):
    if i % 2 == 0:
      total += i
  return total
print(sum_odd(2, 7))
```

```python
def is_palindrome(s):
  for i in range(len(s) // 2):
    if s[i] != s[len(s) - 1 - i]:
      return False
  return True
print(is_palindrome("racecar"))
```

```python
def did_student_pass(score, passing_score=50):
    """
    Checks if a student passed based on their score.

    Parameters:
        score (float or int): The student's score.
        passing_score (float or int): The minimum score required to pass. Default is 50.

    Returns:
        bool: True if the student passed, False otherwise.
    """
    if not isinstance(score, (int, float)):
        raise ValueError("Score must be a number.")

    return score >= passing_score

# Example usage:
print(did_student_pass(75))   # True
print(did_student_pass(45))   # False
```
