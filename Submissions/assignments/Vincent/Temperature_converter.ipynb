```python
# inputs the temperature
celsius =float(input(" Enter temperature value in celsius: "))
#formula to convert the celcius to fahrenheit
fahrenheit =(celsius * 9/5) + 32
print(f"{celsius}째C = {fahrenheit:.2f}째F")

#conditional statement based on celcius range
if celsius < 0:
  print("It is freezing cold")

elif celsius > 0 and celsius <=20:
  print("its a bit chilly")

elif celsius > 21 and celsius <=30:
  print("its nice and warm")

else:
  print("its  hot")

```
