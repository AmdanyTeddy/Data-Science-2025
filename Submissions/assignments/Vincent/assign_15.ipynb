```python
def recur_factorial(n):
    """
    Returns the factorial of n using recursion.
    """
    if n < 0:
        raise ValueError("n must be >= 0")
    if n == 0:
        return 1
    return n * recur_factorial(n - 1)

# ðŸ§ª Test cases:
print(recur_factorial(0))  # âžœ 1
print(recur_factorial(5))  # âžœ 120
print(recur_factorial(7))  # âžœ 5040
```

```python
def is_palindrome_recur(s):
    """
    Returns True if s is a palindrome using recursion, False otherwise.
    """
    if len(s) <= 1:
        return True
    if s[0] != s[-1]:
        return False
    return is_palindrome_recur(s[1:-1])

# ðŸ§ª Test cases:
print(is_palindrome_recur("racecar"))  # âžœ True
print(is_palindrome_recur("hello"))    # âžœ False
print(is_palindrome_recur("madam"))    # âžœ True
```
