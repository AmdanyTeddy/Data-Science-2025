```python
def safe_average(L):
    """
    L is a list of numbers (ints or floats). It may be empty or contain non-numeric values.
    Returns the average of all valid numeric elements in the list.
    Ignores non-numeric items using exception handling.
    Raises a ValueError if there are no numeric elements.
    """
    total = 0
    count = 0

    for item in L:
        try:
            num = float(item)
            total += num
            count += 1
        except (ValueError, TypeError):
            continue

    if count == 0:
        raise ValueError("No numeric values found in the list.")

    return total / count

# ðŸ§ª Example Tests
print(safe_average([10, 20, "thirty", 40]))  # âžœ 23.333...
# print(safe_average(["a", "b"]))            # âžœ raises ValueError
```

```python
def merge_inventory(inv1, inv2):
    """
    Combines two inventories (dictionaries). If items overlap, their quantities are summed.
    Returns a new merged dictionary.
    """
    merged = inv1.copy()  # Start with inv1

    for item, qty in inv2.items():
        if item in merged:
            merged[item] += qty
        else:
            merged[item] = qty

    return merged

# ðŸ§ª Example Tests
inv1 = {"pen": 10, "notebook": 5}
inv2 = {"notebook": 3, "eraser": 7}
print(merge_inventory(inv1, inv2))  # âžœ {'pen': 10, 'notebook': 8, 'eraser': 7}
```
