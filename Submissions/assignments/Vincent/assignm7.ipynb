```python
def keep_consonants():
    word = input("Enter a word: ")
    vowels = "aeiouAEIOU"
    result = ""
    for char in word:
        if char.isalpha() and char not in vowels:
            result += char
    return result

# ðŸ§ª Example usage:
print(keep_consonants())  # If user enters "babas", output will be: bbs
```
